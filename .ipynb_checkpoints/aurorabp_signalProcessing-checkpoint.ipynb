### *Filtragem*

#### **Seleção de features com valores ópticos**
Motivação $\rightarrow$  "Como o PPG é uma técnica $óptica$  que é utilizada para inferir as variações no volume sanguíneo em camadas vasculares, o grupo de atributos ópticos contém maioria de seus indivíduos derivados da onda PPG[...]" (MIELOSZYK et al., 2022, p.5 ) 

````python

optical_features = [col for col in features_df.columns if col.endswith('_optical') and not col.startswith('delta_')]
deltaOpFeatures = [col for col in features_df.columns if col.startswith('delta_') and col.endswith('_optical')]

optical_df = features_df[['pid', 'date_time'] +optical_features].copy()
deltaOp_df = features_df[['pid', 'date_time'] + deltaOpFeatures].copy()
````

# Estudo do sinal

````python

pid_count = 3
osci_count = 2
sampling_rate = 500
ppt_osci_df = oscillometric_df.loc[oscillometric_df.pid == participants_df.pid[pid_count]]
## carregar dataset do waveform
osci_series = ppt_osci_df.iloc[osci_count]
````

````python
osci_series
````

````python

waveform_df = pd.read_csv(os.path.join(data_dir, osci_series.waveform_file_path), delimiter = '\t')
````

````python
# filtro passa faixa
def _bandpass_filter(data: np.ndarray, hp: float, lp: float, fs: float = sampling_rate) -> np.ndarray:
    order = 4
    b, a = signal.butter(order, (hp, lp), btype='bandpass', fs=fs)
    return signal.lfilter(b, a, data-data[0])
````

## *Gráficos de onda*

#### Análise do exemplo em "waveform_example"

````python
#ppg
plt.subplot(3,1,3)
plt.plot(waveform_df.t, _bandpass_filter(waveform_df.optical.to_numpy(), 0.25, 50))
plt.xlim(0,10)
plt.ylabel('PPG (AU)') 
````

#### Gráfico baseado no exemplo

````python
#ppg
fig, ax = plt.subplots(nrows=1, ncols = 1, figsize=(20,5) ) 
ax.patch.set_facecolor('#30332f')

# x = tempo, y = ppg
plt.plot(waveform_df.t, _bandpass_filter(waveform_df.optical.to_numpy(), 0.25, 50), color='#32eb28')

#delimitando a faixa de valores
plt.xlim(0,5)
plt.ylim(-210,220)

#adicionando grades
ax.minorticks_on()
ax.grid(True, which='major' , color='#c4cfc7', linestyle=':', linewidth=0.5, axis='both')         
ax.set_xticks(np.arange(0, 5.1, 0.1))
ax.set_yticks(np.arange(-210,221,30))

#legenda
plt.ylabel('PPG (AU)') 
#pdf
plt.savefig("ppgsign.pdf")
````

### _Análise estatística_

````python
t = waveform_df.t.to_numpy()
optical = waveform_df.optical.to_numpy()

# Defina o intervalo de tempo que deseja filtrar (em segundos)
t_start = 0 
t_end = 5   

# índices correspondentes no intervalo de tempo
idx_start = np.searchsorted(t, t_start, side='left')
idx_end = np.searchsorted(t, t_end, side='right')

# trecho do sinal
optical_range = optical[idx_start:idx_end]

# Aplica o filtro
signal_filtered = _bandpass_filter(optical_range, 0.25, 50, fs=sampling_rate)

# Estatísticas no sinal 
maxi = signal_filtered.max()
mini = signal_filtered.min()
mean = signal_filtered.mean()
std_dev = signal_filtered.std()
amplitude = maxi - mini
````
